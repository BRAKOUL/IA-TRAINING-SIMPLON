# Guide d'analyse et de visualisation

* une variable numerique : histogramme, boxplot, kernel density plot
* une variable catégorielle: piechart, barchat

* 2 variables numériques : scatter plot, heatmap
* 2 variables catégorielles: barchat

* 1 variable numérique, 1 variable catégorielle : Violin plot
* 1 variable numérique , 2 variables catégorielles : Facet grid


### Analyse variables numériques
> **1 variable numérique**
> **=> Analyse `univariée` numérique :** histogramme, boxplot, kernel density plot

> **2 variables numériques**
> **=> Analyse `bivariée` numérique :** scatter plot, heatmap, regplot

> **Plusieurs variables numériques**
> **=> Analyse `multivariée` numérique :** heatmap, pairplot


### Analyse variables catégorielles
> **1 variable catégorielle**
> **=> Analyse `univariée` catégorielle :** piechart, barchat

> **2 variables catégorielles**
> **=> Analyse `bivariée` catégorielle :** barchat

### Analyse (croisée) variables numériques `X` variables catégorielles
> **1 variable numérique, 1 variable catégorielle**
> **=> Analyse `bivariée` :**  Violin plot, Strip plot, Swarm plot

> **1 variable numérique, 2 variables catégorielles**
> **=> Analyse `multivariée` :** Facet grid

> **1 variable numérique , 3 variables catégorielles**
> **=> Analyse `multivariée` :** Catplot (Strip plot + Facet grid)

> **2 variables numériques , 2 variables catégorielles**
> **=> Analyse `multivariée` :** lmplot (Facet grid + Scatter plot)

> **Multiple variables numériques & catégorielles**
> **=> Analyse `multivariée` :** pairplot (matrice de visualisation)
